## Remote Repositories

```bash
# Clone a repo
git clone https://github.com/user/repo.git

# Add remote to existing repo
git remote add origin https://github.com/user/repo.git

# Push changes
git push origin main

# Pull changes
git pull origin main

# Fetch without merging
git fetch origin
```

## Exercise 1: Push to GitHub (10 min)

Practice the push/pull workflow with a real GitHub repository.

1. **Create a new repo on GitHub.** Either use the web interface (github.com → New Repository) or run:
   ```bash
   gh repo create my-test-repo --public --clone
   ```

2. **If using an existing local repo** (e.g., your `git-practice` repo), add the remote:
   ```bash
   git remote add origin https://github.com/<your-username>/my-test-repo.git
   ```

3. **Push your code:**
   ```bash
   git push -u origin main
   ```

4. **Visit your GitHub repo page** — your code is now online! Verify the files are there.

5. **Make an edit on GitHub's web interface:** Click on `README.md` (or any file), click the pencil icon, add a line, and commit directly on GitHub.

6. **Pull the change locally:**
   ```bash
   git pull origin main
   ```

7. **Verify:**
   ```bash
   git log --oneline
   ```
   You should see the commit you made on GitHub.

**Expected outcome:** You have a repo that exists both locally and on GitHub, and you've practiced the push/pull sync cycle.

## Pull Requests

### Workflow

1. Create feature branch
2. Make changes, commit
3. Push branch to GitHub
4. Open Pull Request
5. Review and discuss
6. Merge when approved

```bash
# Create and push branch
git checkout -b feature/add-search
# ... make changes ...
git add .
git commit -m "feat: Add search functionality"
git push -u origin feature/add-search
```

## Exercise 2: Create a Pull Request (10 min)

Practice the full PR workflow on your GitHub repository.

1. **Create a feature branch locally:**
   ```bash
   git checkout -b add-feature
   ```

2. **Make a change.** Add a new file or modify existing code. For example, create `hello.py`:
   ```python
   def greet(name):
       return f"Hello, {name}!"
   ```

3. **Commit and push the branch:**
   ```bash
   git add .
   git commit -m "feat: add greeting function"
   git push -u origin add-feature
   ```

4. **Go to GitHub** and you should see a banner suggesting you create a Pull Request. Click it (or go to the Pull Requests tab → New Pull Request).

5. **Fill in a title and description**, then create the PR.

6. **Explore the PR page.** Answer these questions:
   - What tabs are available (Conversation, Commits, Files changed)?
   - How many commits are in this PR?
   - Can you see the diff of your changes?

**Don't merge yet** — we'll come back to this PR.

**Expected outcome:** You have an open PR on GitHub and understand what information the PR page shows.

## Exercise 3: AI PR Description (5 min)

Use AI to draft a professional PR description.

1. **Ask your AI agent** (Claude, ChatGPT, or your CLI agent):

   > "I'm creating a PR that adds a `calculate_h_index` function to `metrics.py` with 5 pytest tests. Write a PR description."

2. **Read the AI-generated description.** Does it include:
   - A summary of what changed?
   - Why the change was made?
   - How to test it?
   - Any breaking changes?

3. **Compare with what you'd write manually.** Consider:
   - Is the AI version better or worse?
   - What context is the AI missing? (e.g., it doesn't know your project's conventions, related issues, or the motivation behind the feature)
   - Would you use it as-is, or as a starting draft?

**Expected outcome:** You see how AI can accelerate PR descriptions, but also understand its limitations — it can't know *why* you made the change or what issues it relates to.

## Exercise 4: Review a Diff (10 min)

Practice code review — a critical collaboration skill.

**Look at this code change and write review comments:**

```python
# Before:
def search(query):
    r = requests.get(f"https://api.openalex.org/works?search={query}")
    return r.json()

# After:
def search(query, limit=10):
    url = f"https://api.openalex.org/works?search={query}&per_page={limit}" 
    r = requests.get(url)
    r.raise_for_status()
    data = r.json()
    return data['results']
```

1. **Write your own review** (2-3 minutes). Consider:
   - What improved in the new version?
   - What's still missing? (Think about: input validation, docstring, type hints, error handling for JSON parsing, URL encoding of the query string)

2. **Now ask your AI agent to review the same diff:**

   > "Review this code change for bugs, security issues, performance, style, and missing tests: [paste the before/after code]"

3. **Compare your review with the AI's review:**
   - Did the AI catch things you missed?
   - Did you catch things the AI missed?
   - Whose review would be more helpful to the author?

**Expected outcome:** You've practiced reading diffs critically and seen how AI can augment (but not replace) human code review.

## GitHub CLI

```bash
# Install: https://cli.github.com

# Create PR from command line
gh pr create --title "Add search" --body "Description here"

# List PRs
gh pr list

# View PR
gh pr view 123

# Merge PR
gh pr merge 123
```

## Issues

Track bugs and features:

```bash
# Create issue
gh issue create --title "Bug: Search fails" --body "Description"

# List issues
gh issue list

# Close with commit
git commit -m "fix: Search error (closes #123)"
```

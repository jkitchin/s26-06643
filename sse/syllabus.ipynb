## Course Description

This course teaches modern software development practices for scientific research, with a focus on leveraging AI coding agents (Claude Code, GitHub Copilot, etc.) as collaborative tools. Students will learn to build, test, document, and distribute Python packages while developing critical skills in evaluating and directing AI-generated code.

The course balances foundational understanding with practical AI-assisted workflows, preparing students for the evolving landscape of software development.

Topics include:
- AI-assisted software development
- Terminal and shell usage
- Git version control and GitHub collaboration
- Python packaging and distribution
- Testing and code quality
- Debugging and profiling
- CI/CD with GitHub Actions
- Documentation generation
- Configuring and extending AI agents (CLAUDE.md, skills, MCP servers, CLIs)

## Learning Objectives

By the end of this course, students will be able to:

1. Use AI coding agents effectively for software development tasks
2. Critically evaluate AI-generated code for correctness and quality
3. Use version control (Git/GitHub) for collaborative development
4. Create and distribute Python packages with proper structure
5. Write comprehensive tests using AI assistance
6. Debug and profile code systematically with AI assistance
7. Apply code quality tools (linters, formatters, pre-commit hooks)
8. Set up CI/CD pipelines using GitHub Actions
9. Generate documentation with AI assistance
10. Configure and extend AI capabilities with project instructions, skills, and custom tools

### Project

The project is an opportunity to build a complete, publishable Python package using AI assistance throughout. Projects should:
- Have clean package structure (pyproject.toml)
- Include comprehensive tests (>90% coverage)
- Have full documentation (README, API docs, examples)
- Include a CI/CD pipeline (GitHub Actions)
- Be published to TestPyPI or PyPI
- Be presented in a well-documented repository with a reflection document

**Deliverables:**
1. **Proposal** (Week 2): Problem statement, planned approach
2. **Progress Report** (Week 4): Working prototype, challenges encountered
3. **Final Submission** (Week 6): Complete package + reflection document
4. **Presentation** (Week 7): 10-minute demo + Q&A

## Academic Integrity

You are encouraged to discuss concepts with classmates, but all submitted work must be your own. AI assistance is required and encouraged in this course, but you must understand and be able to explain all code you submit. Assignments require honest reflection on AI usage.

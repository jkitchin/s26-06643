## The New Development Paradigm

Software development has fundamentally changed. AI coding agents can:

- Write code from natural language descriptions
- Debug errors and explain fixes
- Generate tests for existing code
- Refactor and improve code quality
- Write documentation
- Explain unfamiliar codebases

**But they are not magic.** They:
- Make mistakes and hallucinate
- Don't truly "understand" code
- Can produce insecure or inefficient code
- Need human guidance and oversight

Your job is to become an effective **human-AI team**.

## What Are LLM Coding Agents?

LLM (Large Language Model) coding agents are AI systems trained on vast amounts of code and text. They can:

### Types of AI Coding Tools

| Tool | Type | Description |
|------|------|-------------|
| **Claude Code** | Agentic | Full terminal access, can run commands, edit files, explore codebases |
| **GitHub Copilot** | Autocomplete | Suggests code as you type in your editor |
| **Cursor** | Hybrid | Editor with AI chat and autocomplete |
| **Cody** | Hybrid | Context-aware coding assistant |
| **ChatGPT** | Chat-based | Conversational code generation |

### Agentic vs. Autocomplete

**Autocomplete tools** (Copilot):
- Suggest the next line or block of code
- Work within your editor
- Require you to drive the development

**Agentic tools** (Claude Code):
- Can execute multi-step tasks autonomously
- Read and write files, run commands
- Explore codebases and make decisions
- Can complete entire features with one prompt

## Effective Prompting Strategies

The quality of AI output depends heavily on your prompts.

### Be Specific

**Bad:**
```
Write a function to process data
```

**Good:**
```
Write a Python function that:
- Takes a list of dictionaries with 'name' and 'score' keys
- Returns the average score
- Raises ValueError if the list is empty
- Include type hints and a docstring
```

### Provide Context

**Bad:**
```
Add error handling
```

**Good:**
```
Add error handling to the fetch_data function in api.py.
It should:
- Catch network errors and retry 3 times
- Log errors using the existing logger
- Raise a custom APIError on failure
```

### Request Specific Formats

```
Create a pytest test file for the Calculator class.
- Use pytest fixtures for setup
- Include edge cases (zero, negative numbers)
- Use parametrize for multiple test cases
- Follow the Arrange-Act-Assert pattern
```

### Iterate and Refine

Don't expect perfection on the first try:

```
> Write a function to parse dates
[AI generates code]

> Good, but also handle ISO format dates
[AI updates code]

> Add error handling for invalid formats
[AI adds try/except]

> Write tests for this function
[AI generates tests]
```

## Critical Evaluation of AI Output

**Never blindly trust AI-generated code.** Always:

### 1. Read and Understand

Before accepting any code:
- Read every line
- Understand what it does
- Ask AI to explain if needed

### 2. Check for Common Issues

| Issue | What to Look For |
|-------|------------------|
| **Security** | SQL injection, command injection, hardcoded secrets |
| **Edge cases** | Empty inputs, None values, boundary conditions |
| **Error handling** | Missing try/except, swallowed exceptions |
| **Performance** | Nested loops, repeated calculations, memory leaks |
| **Dependencies** | Unnecessary imports, version conflicts |
| **Style** | Inconsistent naming, missing docstrings |

### 3. Test Thoroughly

AI-generated code needs testing just like any other code:
- Run the code with sample inputs
- Test edge cases
- Run existing tests
- Add new tests if needed

### 4. Question Hallucinations

AI can confidently generate:
- Nonexistent function names
- Wrong API signatures
- Outdated library versions
- Incorrect algorithms

**If something looks unfamiliar, verify it!**

## Exercise 3: Spot the Issues (10 min)

Practice your critical evaluation skills on AI-generated code.

### Instructions

1. Ask your AI agent:
   ```
   Write a Python function that reads a CSV file and returns the average of the second column.
   ```

### Questions

**(a) What assumptions did the AI make?**
- Does it assume the CSV has headers? 
- Does it assume the second column is numeric?
- Does it assume a specific delimiter?
- Does it assume the file exists?

**(b) What error handling is missing?**
- What happens if the file doesn't exist?
- What happens if the second column has non-numeric values?
- What happens if the CSV has fewer than two columns?
- What happens if the file is empty?

**(c) What would break with unexpected input?**
- A CSV with a header row that isn't numeric
- A CSV that uses semicolons instead of commas
- A CSV with missing values in the second column
- A very large file (millions of rows)

### Discuss

Share your findings with a neighbor. Did you catch the same issues? Did they find something you missed?

**Takeaway:** Reading AI code critically is one of the most important skills in AI-assisted development.

## The Human-AI Collaboration Model

Think of AI as a **very fast junior developer** who:
- Knows a lot of syntax and patterns
- Works incredibly quickly
- Makes confident mistakes
- Needs supervision and guidance

### Your Role

| You Do | AI Does |
|--------|--------|
| Define requirements | Generate code |
| Make design decisions | Implement patterns |
| Review and approve | Write boilerplate |
| Catch errors | Generate tests |
| Understand the system | Document code |
| Own the final product | Suggest improvements |

### When to Use AI

**Great for:**
- Boilerplate and scaffolding
- Standard patterns (CRUD, API calls)
- Test generation
- Documentation
- Explaining code
- Debugging common errors

**Be careful with:**
- Security-critical code
- Complex algorithms
- Domain-specific logic
- Performance-critical code
- Novel problems

## Ethical Considerations

### Attribution

AI-generated code exists in a gray area:
- AI trained on open-source code
- Output may resemble training data
- No clear copyright guidance yet

**Best practice:** Treat AI output as a starting point, not a final product.

### Over-reliance

Risks of depending too heavily on AI:
- Atrophying fundamental skills
- Not understanding your own code
- Missing learning opportunities
- Blind spots in security/performance

**Balance:** Use AI to accelerate, not replace, your learning.

### Environmental Impact

LLMs require significant compute resources:
- Be efficient with prompts
- Don't generate code you won't use
- Consider the carbon footprint

# Weekly Participation Journal

**06-643: Creating Scientific Software with AI Coding Agents**

## Overview

Each week, you will submit a short reflective journal entry documenting your experience working with AI coding agents. This is an ongoing assignment that runs for the duration of the course.

The purpose of this journal is to:
- Build a habit of reflecting on your AI-assisted workflow
- Track your growth and changing strategies over the semester
- Create a personal reference of what works and what doesn't
- Develop critical evaluation skills for AI-generated output

## What to Write

Each weekly entry should be **roughly 200-400 words** and cover:

1. **What AI tools did you use this week?** (Claude Code, Copilot, ChatGPT, Codex, Gemini CLI, Cursor, etc.)
2. **What did you use them for?** (writing code, debugging, explaining concepts, generating tests, writing documentation, etc.)
3. **What worked well?** (effective prompts, good outputs, time saved)
4. **What didn't work?** (hallucinations, wrong answers, unhelpful suggestions, wasted tokens)
5. **What did you learn?** (new concepts, new tools, new prompting strategies, surprises)

You do not need to answer every question every week — focus on what is most relevant. Some weeks you might focus on a single interesting interaction; other weeks you might compare two tools.

## Grading

Participation is 20% of your final grade. You are graded on:
- **Consistency**: Did you submit an entry each week?
- **Thoughtfulness**: Do your entries show genuine reflection, not just surface-level descriptions?
- **Growth**: Do your entries show evolving strategies and deeper understanding over time?

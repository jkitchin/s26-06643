## Part 2: Debug Buggy Code

The following code has **3 bugs**. Use an AI coding agent to help identify and fix them. Document your debugging process.

```python
def analyze_citations(works):
    """Analyze citation patterns in a list of works.

    Each work is a dict with keys: 'title', 'year', 'cited_by_count'
    Returns summary statistics by year.
    """
    yearly_stats = {}
    for work in works:
        year = work['year']
        if year not in yearly_stats:
            yearly_stats[year] = {'count': 0, 'total_citations': 0, 'titles': []}
        yearly_stats[year]['count'] =+ 1
        yearly_stats[year]['total_citations'] += work['cited_by_count']
        yearly_stats[year]['titles'].append(work['title'])

    for year in yearly_stats:
        stats = yearly_stats[year]
        stats['avg_citations'] = stats['total_citations'] / stats['count']
        stats['max_citations'] = max(work['cited_by_count'] for work in works if work['year'] == year)

    sorted_years = sorted(yearly_stats.keys(), reverse=True)
    return {year: yearly_stats[year] for year in sorted_years}


def find_highly_cited(works, threshold=None):
    """Find works cited more than the threshold.

    If threshold is None, use the mean citation count.
    """
    if threshold is None:
        threshold = sum(w['cited_by_count'] for w in works) / len(works)

    result = []
    for work in works:
        if work['cited_by_count'] > threshold:
            result.append(work)

    result.sort(key=lambda w: w['cited_by_count'])
    return result
```

# Module 02: Git and Version Control

Track every change with confidence.

## Learning Objectives

1. Understand why version control is essential
2. Initialize and manage Git repositories
3. Create commits, branches, and merges
4. Use AI to explain Git concepts and resolve errors
5. Develop good commit habits

## Why Version Control?

Especially important with AI-generated code:

- **Track changes**: See what AI modified
- **Rollback**: Undo bad AI suggestions
- **Branches**: Experiment safely
- **Collaboration**: Work with others (and AI)
- **History**: Understand how code evolved

**Rule**: Never accept AI code without committing first!

## Git Basics

### Configuration

```bash
# Set your identity
git config --global user.name "Your Name"
git config --global user.email "you@example.com"

# Set default branch name
git config --global init.defaultBranch main

# Check config
git config --list
```

### Initialize a Repository

```bash
# Create new repo
git init

# Clone existing repo
git clone https://github.com/user/repo.git
```

## The Git Workflow

```
Working Directory → Staging Area → Repository
     (edit)           (add)         (commit)
```

### Basic Commands

```bash
# Check status
git status

# Stage changes
git add file.py           # Specific file
git add .                  # All changes
git add -p                 # Interactive staging

# Commit
git commit -m "Add feature X"

# View history
git log
git log --oneline
git log --graph --oneline
```

### AI Help for Git

```
> I made changes but want to undo them before committing

Claude: Use git restore:
git restore file.py           # Discard changes to specific file
git restore .                  # Discard all changes
git restore --staged file.py  # Unstage a file
```

## Writing Good Commit Messages

### Format

```
type: Short summary (50 chars or less)

Longer description if needed. Wrap at 72 characters.
Explain what and why, not how.
```

### Types

| Type | Description |
|------|-------------|
| `feat` | New feature |
| `fix` | Bug fix |
| `docs` | Documentation |
| `test` | Adding tests |
| `refactor` | Code restructuring |
| `style` | Formatting |
| `chore` | Maintenance |

## Branching

Branches let you develop features in isolation.

```bash
# Create and switch to branch
git checkout -b feature-name
# or (modern)
git switch -c feature-name

# List branches
git branch

# Switch branches
git checkout main
git switch main

# Delete branch
git branch -d feature-name
```

### Merging

```bash
# Merge feature into main
git checkout main
git merge feature-name
```

### AI Help with Merge Conflicts

```
> I have a merge conflict in file.py. Here's the conflict markers: [shows file]

Claude: The conflict shows two versions:
- Lines between <<<<<<< HEAD and ======= are from main
- Lines between ======= and >>>>>>> feature are from your branch

To resolve, choose which version to keep (or combine them),
remove the conflict markers, then:
git add file.py
git commit
```

## Exercise 4: Create and Resolve a Conflict (10 min)

Merge conflicts are inevitable — let's create one on purpose so you know how to handle it.

1. In your `git-practice` repo, make sure you're on `main`:
   ```bash
   git checkout main
   ```
2. Edit `README.md` to add a second line: `"This is the main branch"`. Then commit:
   ```bash
   echo "This is the main branch" >> README.md
   git add README.md
   git commit -m "docs: Add branch info to README"
   ```
3. Create and switch to a new branch:
   ```bash
   git checkout -b conflict-demo
   ```
4. Edit `README.md` — **replace** the second line with `"This is the feature branch"`. Then commit:
   ```bash
   # Edit the file (use nano, vim, or your editor)
   git add README.md
   git commit -m "docs: Update branch info in README"
   ```
5. Switch back to main and merge:
   ```bash
   git checkout main
   git merge conflict-demo
   ```
6. **You should see a merge conflict!** Open `README.md` — you'll see conflict markers:
   ```
   <<<<<<< HEAD
   This is the main branch
   =======
   This is the feature branch
   >>>>>>> conflict-demo
   ```
7. **Resolve it:** Edit the file to keep the content you want (or combine both lines). Remove all conflict markers (`<<<<<<<`, `=======`, `>>>>>>>`).
8. Complete the merge:
   ```bash
   git add README.md
   git commit -m "fix: Resolve merge conflict in README"
   ```

**If you get stuck:** Ask your AI agent — *"How do I resolve a merge conflict in git?"*

**Key takeaway:** Conflicts look scary but are just Git asking you to make a decision. The markers show you both versions so you can choose.

## Viewing Changes

```bash
# See uncommitted changes
git diff

# See staged changes
git diff --staged

# See changes in a commit
git show abc1234

# See file at specific commit
git show abc1234:path/to/file.py

# Compare branches
git diff main..feature-branch
```

### Understanding Diffs

```diff
- removed line (red)
+ added line (green)
  unchanged context
```

## Undoing Things

### Before Committing

```bash
# Discard changes
git restore file.py

# Unstage
git restore --staged file.py
```

### After Committing

```bash
# Amend last commit (change message or add files)
git commit --amend

# Undo last commit but keep changes
git reset --soft HEAD~1

# Undo last commit and discard changes
git reset --hard HEAD~1

# Create a new commit that undoes a previous commit
git revert abc1234
```

### AI Help for Recovery

```
> I accidentally committed sensitive data. How do I remove it from history?

Claude: WARNING: This rewrites history. If pushed, coordinate with team.

git filter-branch --force --index-filter \
  "git rm --cached --ignore-unmatch path/to/file" \
  --prune-empty --tag-name-filter cat -- --all

Then force push: git push --force
```

## Tips and Tricks

- **Let AI write your commit messages**: Prompt "Write a commit message for these changes: [paste diff]" saves time and produces clearer, more consistent messages.
- **Commit early, commit often**: Small commits are easier to review, easier to revert, and give you more rollback points when experimenting with AI-generated changes.
- **Commit before letting the agent work**: Always commit your current state before asking the agent to make changes. Then use `git diff` to see exactly what it did.
- **Use `git stash` before experimental sessions**: If things go wrong during an AI session, `git stash pop` gets you back to a clean state instantly.
- **Ask AI to explain git history**: Prompt "Explain what this git log shows: [paste log]" is faster than decoding complex branch histories yourself.
- **Guard dangerous git operations**: Never let the agent run `git push --force` or `git reset --hard` without understanding why. These can destroy work permanently.
- **Generate `.gitignore` early**: Ask "Generate a .gitignore for a Python project" at the start. This prevents committing virtual environments, cache files, and other junk.
- **Ask AI to diagnose confusing state**: If you are confused, prompt "Run git status and git log --oneline -10 and explain what you see." This is faster than puzzling through it yourself.

## Foundational Concepts to Commit to Memory

- **Working directory, staging area, repository**: These are the three states of your files. You edit in the working directory, `git add` moves changes to staging, and `git commit` saves them permanently. Know where your changes are at all times.
- **Commits are snapshots, not diffs**: Each commit records the full state of your project at that moment. This is why you can always go back to any previous state.
- **Branches are cheap pointers**: A branch is just a label pointing to a commit. Creating a branch costs almost nothing. Use them freely to isolate work.
- **Conventional commit format**: `type: description` where type is one of `feat`, `fix`, `docs`, `test`, `refactor`, `style`, `chore`. This is not optional style -- it is how teams communicate through history.
- **git status is your best friend**: Run it constantly. It tells you what branch you are on, what is staged, what is modified, and what is untracked. When in doubt, check status.
- **Merge conflicts are decisions, not errors**: When Git shows conflict markers, it is asking you to choose. Read both versions, decide what the code should be, remove the markers, and commit.
- **git diff before git add**: Always review what you are about to stage. This is especially important with AI-generated changes where you might not have seen every modification.
- **.gitignore prevents accidents**: Set it up at the start of every project. Never commit secrets, build artifacts, or environment-specific files.

## Knowing vs. Doing: Reflection

Git is a tool you can use productively after learning a handful of commands, but it is also a tool you can study for years and still discover new capabilities. That tension is worth sitting with. You do not need to understand the internals of Git's directed acyclic graph to make commits, create branches, and resolve conflicts. But the more you understand about how Git actually works -- that branches are pointers, that commits are snapshots, that the reflog tracks everything -- the faster you can recover from mistakes and the more confidently you can work.

AI agents are particularly good at Git. They can generate commit messages, explain diffs, resolve merge conflicts, and construct complex `git log` queries you would never memorize. This is a genuine superpower: you can ask "how do I undo the last three commits but keep my changes?" and get an accurate answer in seconds instead of searching Stack Overflow for twenty minutes. The AI expands your effective Git knowledge well beyond what you have memorized, and that lets you move faster.

But here is the catch: Git mistakes can be consequential. A bad `git reset --hard` can destroy work. A force push can overwrite a teammate's commits. The value you bring is not in knowing every Git command -- it is in understanding what operations are safe and what operations are destructive, in knowing when to branch before experimenting, and in building the habit of committing before you let AI change your code. Those instincts only come from practice and from understanding the concepts well enough to know what questions to ask. Let AI handle the syntax. You handle the judgment.

## Docstrings

Google style:
```python
def search_works(query: str, limit: int = 10) -> list[dict]:
    """Search OpenAlex for scholarly works.
    
    Args:
        query: Search query string.
        limit: Maximum number of results.
    
    Returns:
        List of work dictionaries with title, year, citations.
    
    Raises:
        ValueError: If query is empty.
        APIError: If the API request fails.
    
    Examples:
        >>> search_works("machine learning", limit=5)
        [{'title': '...', 'year': 2024, ...}, ...]
    """
```

> **Context Engineering Reminder** (from Module 00)
>
> Before starting this exercise, think about what context the agent needs. For documentation, that means:
> - Have the agent read the source code before writing docstrings — it needs to see parameter types, return values, and logic
> - Show it an example of your preferred docstring style (numpydoc, Google, Sphinx) or point to an existing docstring in the project
> - For READMEs, load `pyproject.toml` and a sample usage script so the agent can describe accurate installation and usage
>
> Load the context first, then ask for the action.

## Exercise 1: Write a Docstring (5 min)

Practice writing a Google-style docstring by hand before using AI.

1. **Write a docstring for this function** (without AI help first):

   ```python
   def filter_by_year(works, start_year, end_year=None):
       # filters works list to only include those published
       # between start_year and end_year (inclusive)
       ...
   ```

   Include all of these sections:
   - **Summary** (one line)
   - **Args** (with types for `works`, `start_year`, `end_year`)
   - **Returns** (what type and what it contains)
   - **Raises** (what errors could occur?)
   - **Example** (a short usage example)

2. **Now ask your AI agent** to write a docstring for the same function.

3. **Compare the two.** Consider:
   - Whose is more accurate?
   - Did the AI assume things about the function that might be wrong?
   - Did you forget any sections that the AI included?

**Expected outcome:** You can write a complete Google-style docstring from memory, and you understand where AI drafts need human verification.

## Exercise 2: AI Docstring Audit (10 min)

Use AI to draft docstrings, then critically review them — a realistic professional workflow.

1. **Pick 3 functions** from your course project (or from any Python module you've written). Choose functions that currently lack docstrings.

2. **Ask your AI agent** to add docstrings to all of them:

   > "Add Google-style docstrings to these functions: [paste your code]"

3. **Review each generated docstring.** For each one, check:
   - Is the description accurate? (AI sometimes guesses wrong about what a function does)
   - Are the parameter types correct?
   - Are the return types correct?
   - Are the examples actually runnable? (Try them!)
   - Is anything important missing?

4. **Fix any issues** you find. Common AI mistakes:
   - Incorrect type annotations (e.g., saying `list` when it's actually `dict`)
   - Examples that don't match the actual function signature
   - Missing edge cases in the description

**Expected outcome:** You have 3 functions with reviewed, accurate docstrings. You've practiced the "AI drafts, human reviews" workflow that's standard in industry.

## Tips and Tricks

- **Prompt: "Write a numpydoc docstring for [function]"**: AI generates well-formatted docstrings when you specify the style (numpydoc, Google, Sphinx).
- **Document as you go, not at the end**: Ask the agent to add docstrings immediately after writing a function — it is faster when the context is fresh.
- **Prompt: "Write a README for this project based on [pyproject.toml and source files]"**: AI can draft a solid README from your project metadata.
- **Use examples in docstrings**: "Add a doctest example to this docstring" — examples are the most useful part of documentation for readers.
- **Prompt: "Generate API documentation for all public functions in [module]"**: AI can batch-generate docstrings for an entire module in one pass.
- **Build docs locally before pushing**: `jupyter-book build .` or `sphinx-build . _build` — catch broken links and formatting issues early.
- **Prompt: "Write a tutorial notebook that demonstrates [feature]"**: AI generates good tutorial-style content when you specify the audience.
- **Keep your README under one screen of scrolling**: If it is longer, split into separate docs pages — a README should orient, not teach.

## Foundational Concepts to Commit to Memory

- **Docstrings are documentation** = they are not just comments; tools like Sphinx, pdoc, and MkDocs extract them automatically to build your API reference
- **Google-style docstring sections** = Summary, Args, Returns, Raises, Examples — know this order by heart
- **Type hints + docstrings** = type hints tell tools and readers what types to expect; docstrings explain what things mean and how to use them; you need both
- **README is your front door** = most people decide whether to use your package in under 30 seconds based on the README alone
- **README essentials** = project name, one-line description, installation, quick usage example, license — anything less and users will move on
- **Sphinx vs. MkDocs** = Sphinx is the traditional Python documentation tool (reStructuredText-based); MkDocs is simpler and Markdown-based; both auto-generate docs from docstrings
- **"AI drafts, human reviews"** = the standard industry workflow for documentation; AI is excellent at producing first drafts of docstrings and READMEs, but only you know if they are accurate
- **No docstring = no documentation** = if you skip the docstring, auto-generated docs will show an empty entry; this is the strongest practical argument for writing them

## Knowing vs. Doing: Reflection

Documentation is perhaps the clearest example of where AI agents change the economics of software development. Writing docstrings and READMEs used to be pure overhead — necessary but slow, and most people skipped it because the cost felt too high relative to the immediate benefit. Now AI can draft a complete set of docstrings for a module in seconds. The bottleneck has shifted from writing to reviewing. And reviewing requires knowledge: you need to know what a good docstring looks like, what information matters to the reader, and whether the AI's description of your function is actually accurate.

This is a pattern you will see throughout your career. AI keeps lowering the cost of producing things — code, docs, tests, analyses — but it raises the importance of knowing enough to evaluate what was produced. If you cannot tell whether a docstring is accurate, the AI draft is worse than no docstring at all, because it gives false confidence. The more you understand about your own code and about documentation conventions, the faster you can review and approve (or fix) what AI generates. Knowledge is not just for doing the work yourself; it is for quality control.

The trap to avoid is thinking you need to master Sphinx configuration, reStructuredText syntax, and every documentation framework before you write a single docstring. You do not. Start with Google-style docstrings and a clean README. Ship that. Learn Sphinx or MkDocs when you actually need a documentation website. AI can help you set up those tools when the time comes — but only if you know enough to tell it what you want. There is always more to learn about documentation, and there is always more documentation to write. Focus on the documentation that helps someone use your code today.

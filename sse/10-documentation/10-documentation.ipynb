## Docstrings

Google style:
```python
def search_works(query: str, limit: int = 10) -> list[dict]:
    """Search OpenAlex for scholarly works.
    
    Args:
        query: Search query string.
        limit: Maximum number of results.
    
    Returns:
        List of work dictionaries with title, year, citations.
    
    Raises:
        ValueError: If query is empty.
        APIError: If the API request fails.
    
    Examples:
        >>> search_works("machine learning", limit=5)
        [{'title': '...', 'year': 2024, ...}, ...]
    """
```

## Exercise 1: Write a Docstring (5 min)

Practice writing a Google-style docstring by hand before using AI.

1. **Write a docstring for this function** (without AI help first):

   ```python
   def filter_by_year(works, start_year, end_year=None):
       # filters works list to only include those published
       # between start_year and end_year (inclusive)
       ...
   ```

   Include all of these sections:
   - **Summary** (one line)
   - **Args** (with types for `works`, `start_year`, `end_year`)
   - **Returns** (what type and what it contains)
   - **Raises** (what errors could occur?)
   - **Example** (a short usage example)

2. **Now ask your AI agent** to write a docstring for the same function.

3. **Compare the two.** Consider:
   - Whose is more accurate?
   - Did the AI assume things about the function that might be wrong?
   - Did you forget any sections that the AI included?

**Expected outcome:** You can write a complete Google-style docstring from memory, and you understand where AI drafts need human verification.

## Exercise 2: AI Docstring Audit (10 min)

Use AI to draft docstrings, then critically review them — a realistic professional workflow.

1. **Pick 3 functions** from your course project (or from any Python module you've written). Choose functions that currently lack docstrings.

2. **Ask your AI agent** to add docstrings to all of them:

   > "Add Google-style docstrings to these functions: [paste your code]"

3. **Review each generated docstring.** For each one, check:
   - Is the description accurate? (AI sometimes guesses wrong about what a function does)
   - Are the parameter types correct?
   - Are the return types correct?
   - Are the examples actually runnable? (Try them!)
   - Is anything important missing?

4. **Fix any issues** you find. Common AI mistakes:
   - Incorrect type annotations (e.g., saying `list` when it's actually `dict`)
   - Examples that don't match the actual function signature
   - Missing edge cases in the description

**Expected outcome:** You have 3 functions with reviewed, accurate docstrings. You've practiced the "AI drafts, human reviews" workflow that's standard in industry.

## README

Essential sections:

```markdown
# Package Name

Brief description.

## Installation

```bash
pip install package-name
```

## Quick Start

```python
from package import function
result = function("example")
```

## Features

- Feature 1
- Feature 2

## License

MIT
```

Ask AI to generate from your code!

## Exercise 3: README from Scratch (10 min)

Write a concise, effective README for your project.

1. **Create (or rewrite) your project's `README.md`** with these sections:
   - **Project title**
   - **One-line description** (what does it do?)
   - **Installation instructions** (`pip install ...` or `git clone ...`)
   - **Quick usage example** (3-5 lines of working Python code)
   - **Link to documentation** (if you have any, or "Coming soon")

2. **Keep it under 30 lines.** A good README is concise. You can always add more later.

3. **Ask your AI agent to review it:**

   > "Is this README clear enough for a new user who has never seen my project? What's missing?"

4. **Revise based on the feedback.** Common things people forget:
   - What Python version is required?
   - What are the dependencies?
   - A real, copy-pasteable example

**Expected outcome:** You have a clean, professional README that a stranger could follow to install and use your package.

## API Documentation

Use Sphinx or MkDocs to generate docs from docstrings:

```bash
pip install sphinx sphinx-autodoc-typehints
sphinx-quickstart docs
sphinx-build docs docs/_build
```

Or with MkDocs:
```bash
pip install mkdocs mkdocstrings[python]
mkdocs new .
mkdocs serve
```

## Exercise 4: Generate API Docs (10 min)

See what auto-generated documentation looks like for your code.

**Option A: Using pdoc (quick and easy)**

1. Install pdoc:
   ```bash
   pip install pdoc
   ```

2. Generate HTML docs for your package:
   ```bash
   pdoc your_package_name --html -o docs/
   ```
   (Replace `your_package_name` with your actual package name)

3. Open the generated HTML file in your browser. Explore it:
   - Are all your functions listed?
   - Do functions without docstrings look sparse?
   - How do the docstrings you wrote in Exercise 2 look in the rendered docs?

**Option B: Using AI (if pdoc doesn't work for your setup)**

1. Ask your AI agent:

   > "Generate markdown API documentation for these functions: [paste your module code]"

2. Review the output. Is it accurate?

**Reflect:** If your functions lack docstrings, the generated docs will be nearly empty. This is the strongest motivation for writing good docstrings — they become your documentation automatically.

**Expected outcome:** You've seen auto-generated API docs and understand the direct connection between docstrings and documentation quality.
